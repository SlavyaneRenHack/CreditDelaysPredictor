Добавим колонку - количество месяцев до окончания кредита

```python
current_date = pd.Timestamp.now()

def mounth_until_end(open_date, term):
    mounth_passed = (current_date.year - open_date.year) * 12 + current_date.month - open_date.month
    reamin_mounths = term - mounth_passed
    return max(0, reamin_mounths)

credit_ds["REAMIN_MOUNTHS"] = credit_ds.apply(lambda row: mounth_until_end(row['VALUE_DT'], row['TERM']), axis=1)
```

Заметим, что в колонке PRODUCT_CODE есть 3 основных типа:
- blockbuster
- grace145
- BB

Объединим в 4 категории

```python
card_ds["PRODUCT_CODE"].value_counts()
```

```python
def replace_product_code(value):
    value = value.lower()
    if 'grace' in value:
        return 'grace'
    elif 'blockbuster' in value:
        return 'blockbuster'
    elif '_bb_' in value:
        return 'bb'

    return 'other'

card_ds["PRODUCT_CODE"] = card_ds["PRODUCT_CODE"].apply(replace_product_code)
```

## Создаем сводную таблицу
### Новые колонки
Создадим две модели (а соответственно и два датасета): один для определения вероятности просрочки по кредиту, а другая - по кредитной карте. Создадим сначала один общий датасет, на основе которого будут собраны два более узких.

Добавим следующие колонки:
- количество погашенных кредитов
- количество активных кредитов
- максимальный срок активного кредита
- общая сумма активных кредитов
- количество просрочек по кредитам
- количество кредитнх карт
- количество просрочек по кредиткам

```python
df_base = client_ds.copy()
df_base.head()
```

```python
df_paid_off_counts = credit_ds[credit_ds["REAMIN_MOUNTHS"] == 0].groupby("CLIENT_ID").size().reset_index(name="paid_off_count")
df_active_counts = credit_ds[credit_ds["REAMIN_MOUNTHS"] > 0].groupby("CLIENT_ID").size().reset_index(name="active_count")
df_active_sums = credit_ds[credit_ds["REAMIN_MOUNTHS"] > 0].groupby("CLIENT_ID")["ORIG_AMOUNT"].sum().reset_index(name="active_sum")
df_max_term = credit_ds.groupby('CLIENT_ID')['TERM'].max().reset_index(name='max_term')
df_delay_counts = credit_ds[credit_ds['delay'] > 0].groupby("CLIENT_ID").size().reset_index(name="credits_delay_count")

df_cards_count = card_ds.groupby('CLIENT_ID').size().reset_index(name="cards_count")
df_cards_delay_count = card_ds[card_ds['delay'] > 0].groupby('CLIENT_ID').size().reset_index(name="cards_delay_count")
```

```python
df_base = pd.merge(df_base, df_paid_off_counts, on='CLIENT_ID', how='left')
df_base = pd.merge(df_base, df_active_counts, on='CLIENT_ID', how='left')
df_base = pd.merge(df_base, df_active_sums, on='CLIENT_ID', how='left')
df_base = pd.merge(df_base, df_max_term, on='CLIENT_ID', how='left')
df_base = pd.merge(df_base, df_delay_counts, on='CLIENT_ID', how='left')

df_base = pd.merge(df_base, df_cards_count, on='CLIENT_ID', how='left')
df_base = pd.merge(df_base, df_cards_delay_count, on='CLIENT_ID', how='left')
```

```python
df_base['paid_off_count'] = df_base['paid_off_count'].fillna(0).astype('int64')
df_base['active_count'] = df_base['active_count'].fillna(0).astype('int64')
df_base['max_term'] = df_base['max_term'].fillna(0).astype('int64')
df_base['active_sum'] = df_base['active_sum'].fillna(0)
df_base['credits_delay_count'] = df_base['credits_delay_count'].fillna(0).astype('int64')

df_base['cards_count'] = df_base['cards_count'].fillna(0).astype('int64')
df_base['cards_delay_count'] = df_base['cards_delay_count'].fillna(0).astype('int64')
```
